```python
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
MODEL = 'gpt-4o-mini'
openai = OpenAI()
```

```python
chatbot_system_prompt = """You are a highly intelligent and knowledgeable technical support assistant.
You have expertise in programming, software development, and computer science.
You provide clear, concise, and accurate answers to technical questions with detailed explanations and 
examples when appropriate."""

def get_chatbot_user_prompt(question):
    return f"""You are a technical support assistant. Please answer the following question in 
    detail with explanations and examples where appropriate.{question}"""
question = str(input("Please enter your technical question: "))

response = openai.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": chatbot_system_prompt},
        {"role": "user", "content": get_chatbot_user_prompt(question)}
    ],
    stream=True
)
# Display the streaming response
collected_response = ""
display_handle = display(display_id=True)

for chunk in response:
    if chunk.choices[0].delta.content:
        content = chunk.choices[0].delta.content
        collected_response += content
        display_handle.update(Markdown(collected_response))
```
